This part we are working on the [WHO](http://apps.who.int/gho/data/node.main.A858?lang=en) dataset downloaded from WHO website. And the data including ratio of country population on high blood pressure(blood pressure over SBP140), cholesterol on two levels(190mg/dl for a little bit high and 240mg/dl for very high cholesterol), BMI(BMI>25 for overweight and BMI>30 for obesity) and also death rate of noncommunicable diseases(including heart disease, stroke, cancer, diabetes and chronic lung disease). And we choose statistics data in the same year as possible.

The output data is in aggregation of risk factors and sex by country, intended to estimate regional specific public health condition. And the ratio is calculated after age-standardized (per 100 000 population). Age standardization  is used to allow populations to be compared when the age profiles of the populations are quite different. Considering huge difference in population distribution worldwide, we decided to work on age standardized data rather than a crude one.

```python
import pandas as pd
import numpy as np
```

```python
FILE_FOLDER = './dataset/WHO/'
```

```python
BP_2015 = pd.read_csv(FILE_FOLDER+'BloodPressureSBP140.csv',skiprows=3).iloc[:,0:4]
BP_2015.head()
```

```python
BP_2015.count()
```

```python
# age-standardized estimate
idx = [0,3,4,5]
Chol_2008_190 = pd.read_csv(FILE_FOLDER+'Cholesterol190ByCountry.csv',skiprows=1).iloc[:,idx]
Chol_2008_190.head()
```

```python
Chol_2008_190.count()
```

```python
idx = [0,3,4,5]
Chol_2008_240 = pd.read_csv(FILE_FOLDER+'Cholesterol240ByCountry.csv',skiprows=1).iloc[:,idx]
Chol_2008_240.head()
```

```python
Chol_2008_240.count()
```

```python
idx = [0,4,5,6]
BMI_2015_25 = pd.read_csv(FILE_FOLDER+'NCD_BMI_25A.csv',skiprows=3).iloc[:,idx]
BMI_2015_25.columns=['Country','Both sexes','Male','Female']
BMI_2015_25.head()
```

```python
BMI_2015_25.count()
```

```python
idx = [0,4,5,6]
BMI_2015_30 = pd.read_csv(FILE_FOLDER+'NCD_BMI_30A.csv',skiprows=3).iloc[:,idx]
BMI_2015_30.columns=['Country','Both sexes','Male','Female']
BMI_2015_30.head()
```

```python
BMI_2015_30.count()
```

```python
NCD = pd.read_csv(FILE_FOLDER+'NoncommunicableDiseasesByCountry.csv',skiprows=2).iloc[:,0:5]
```

```python
NCD_2015 = NCD.loc[NCD.Year==2015].copy().reset_index().drop(columns=['Year','index'])
```

```python
NCD_2015.head()
```

```python
NCD_2015.count()
```

```python
Chol = pd.merge(Chol_2008_240,Chol_2008_190,how='outer',on='Country',suffixes=('_chol240','_chol190'))
```

```python
BMI = pd.merge(BMI_2015_25,BMI_2015_30,how='outer',on='Country',suffixes=('_BMI25','_BMI30'))
```

```python
tmp1 = pd.merge(Chol,BMI,how='outer',on='Country',suffixes=('',''))
```

```python
tmp2 = pd.merge(NCD_2015,BP_2015,how='outer',on='Country',suffixes=('_NCD','_BP'))
```

```python
WHOdata = pd.merge(tmp1,tmp2,how='outer',on='Country',suffixes=('',''))
WHOdata
```

```python
WHOdata.to_csv('./SavedList/WHOdata.csv')
```
