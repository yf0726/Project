**You don't need to run this part. It is just for showing how I get the TotalHtml2Page.csv.**

```python
cnt = 0
new_website = []
name_html = []
null = []
for root, _, files in walk(File_Folder):   #return three list 
    for name in files:
        cnt = cnt + 1
        if cnt%500 == 0:
            print('=================================')
            print('Already done:',cnt/len(files)*100,'%')
            print('=================================')
        soup_i = BeautifulSoup(open(File_Folder + name, 'rb'), 'html.parser')
        if soup_i.findAll('a'):
            comments=soup_i.find_all(string=lambda text:isinstance(text,Comment))
            new = re.search(r'http://(.*?)/', comments[0].split()[1]).group(1)
            new_website.append(new)
            name_html.append(name)
            print(new,comments[0])
        else:
            null.append(name)
            print('null',name)
```

```python
new_website.reverse()
name_html.reverse()
```

```python
pd_html = pd.DataFrame([new_website,name_html]).T
```

```python
pd_html.columns = ['website','htmlfile']
```

```python
pd_html.to_csv('TotalHtml2Page.csv',index=False)
```
